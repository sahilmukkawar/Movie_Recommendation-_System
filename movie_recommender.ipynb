```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head().shape
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies['genres']
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
       L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast
def convert5(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if(count!=3):
          L.append(i['name'])
          count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert5)
```

```python
movies.head()
```

```python
import ast
def fetch_data(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
           L.append(i['name'])
           break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_data)
```

```python
movies.head()
```

```python
movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies.head()
```

```python
# Convert each column to a string and concatenate them
movies['tags'] = movies['overview'].apply(str) + " " + \
                 movies['genres'].apply(lambda x: " ".join(x) if isinstance(x, list) else "") + " " + \
                 movies['keywords'].apply(lambda x: " ".join(x) if isinstance(x, list) else "") + " " + \
                 movies['cast'].apply(lambda x: " ".join(x) if isinstance(x, list) else "") + " " + \
                 movies['crew'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
```

```python
sigma = movies[['movie_id', 'title', 'tags']]
```

```python
sigma.head()
```

```python
sigma['tags']=sigma['tags'].apply(lambda x:x.lower())
```

```python
sigma.head(1000)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
sigma['tags']=sigma['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(sigma['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = sigma[sigma['title']== movie].index[0]
    dis=similarity[movie_index]
    movie_list= sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(sigma.iloc[i[0]].title)
```

```python
recommend('Star Trek: Insurrection')
```
